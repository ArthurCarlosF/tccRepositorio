## 1.0 Iniciando rede neural preditiva pré treinada

```python

class redeNeuralPreditiva(nn.Module):
    tamanhoEntrada=105
    tamanhoSaida=2
    def __init__(self):
        super(redeNeuralPreditiva, self).__init__()
          #define a arquitetura
        self.linear1=nn.Linear(self.tamanhoEntrada,6000)
        self.act1=nn.ReLU6()

        self.linear2=nn.Linear(6000,3000)
        self.act2=nn.ReLU6()

        self.linear3=nn.Linear(3000,1600)
        self.act3=nn.ReLU6()
        
        self.linear4=nn.Linear(1600,800)
        self.act4=nn.ReLU6()
        
        self.linear5=nn.Linear(800,400)
        self.act5=nn.ReLU6()

        self.linear6=nn.Linear(400,200)
        self.act6=nn.ReLU6()

        self.linear7=nn.Linear(200,100)
        self.act7=nn.ReLU6()
        
        self.linear8=nn.Linear(100,10)
        self.act8=nn.ReLU6()
        
        self.linear9=nn.Linear(10,self.tamanhoSaida)
        self.act9=nn.Sigmoid()
        
        
        self.melhorErro=1000000
        
    def forward(self, X):
       

        v1=self.linear1(X)
        a1=self.act1(v1)
        
        v2=self.linear2(a1)
        a2=self.act2(v2)
        
        v3=self.linear3(a2)
        a3=self.act3(v3)
        
        v4=self.linear4(a3)
        a4=self.act4(v4)
        
        v5=self.linear5(a4)
        a5=self.act5(v5)
        
        v6=self.linear6(a5)
        a6=self.act6(v6)
        
        v7=self.linear7(a6)
        a7=self.act7(v7)
        
        v8=self.linear8(a7)
        a8=self.act8(v8)
        
        
        vS=self.linear9(a8)
        yh=self.act9(vS)
        return yh
    
    
print("Iniciando rede neural preditiva ...")
roboQuadrupede=redeNeuralPreditiva()

roboQuadrupede.load_state_dict(torch.load("parametrosRedePreditiva.pt"))
roboQuadrupede.eval()
file1 = open("melhorErroPreditiva.txt", "r")  # append mode
roboQuadrupede.melhorErro=float(file1.read())
file1.close()
print("Rede neural preditiva iniciada")
    
```

## 2.0 Criando Controlador Neural



```python
class ControladorNeural(nn.Module):
    tamanhoEntrada=4
    tamanhoSaida=12
    def __init__(self):
        super(ControladorNeural, self).__init__()
          #define a arquitetura
        self.linear1=nn.Linear(self.tamanhoEntrada,4)
        self.act1=nn.ReLU6()

        self.linear2=nn.Linear(4,2)
        self.act2=nn.ReLU6()

        self.linear3=nn.Linear(2,4)
        self.act3=nn.ReLU6()
        
        self.linear4=nn.Linear(4,8)
        self.act4=nn.ReLU6()
        
        self.linear5=nn.Linear(8,9)
        self.act5=nn.ReLU6()

        self.linear6=nn.Linear(9,10)
        self.act6=nn.ReLU6()

        self.linear7=nn.Linear(10,11)
        self.act7=nn.ReLU6()
        
        self.linear8=nn.Linear(11,12)
        self.act8=nn.ReLU6()
        
        self.linear9=nn.Linear(12,self.tamanhoSaida)
        self.act9=nn.Sigmoid()
        
        
        self.melhorErro=1000000
        
    def forward(self, X):
       

        v1=self.linear1(X)
        a1=self.act1(v1)
        
        v2=self.linear2(a1)
        a2=self.act2(v2)
        
        v3=self.linear3(a2)
        a3=self.act3(v3)
        
        v4=self.linear4(a3)
        a4=self.act4(v4)
        
        v5=self.linear5(a4)
        a5=self.act5(v5)
        
        v6=self.linear6(a5)
        a6=self.act6(v6)
        
        v7=self.linear7(a6)
        a7=self.act7(v7)
        
        v8=self.linear8(a7)
        a8=self.act8(v8)
        
        
        vS=self.linear9(a8)
        yh=self.act9(vS)
        return yh
```

```python
setPoint=0,0




from matplotlib import pyplot as plt
import time
tic=time.time()
tempoSimulacao=5 #define o tempo de simulacao em segundos
  
X = []
Y = []
graficoErro=[]

xAtual=0
yAtual=0

##Construindo a entrada inicial
estadoInicial=[]

qAcertos=0
qErros=0

for k in range(3):
    for j in EM:
        estadoInicial.append(j)
    for j in CM:
        estadoInicial.append(j)
    for j in Gy:
        estadoInicial.append(j)
    for j in Ac:
        estadoInicial.append(j)
    estadoInicial.append(Tmp)
    estadoInicial.append(ult)
    for j in posicao:
        estadoInicial.append(j)

estadoAtual=estadoInicial

while((time.time()-tic)<tempoSimulacao):
    
    saida=""
    
    delay=1
    
    #Passo 0: Exibe o estado atual e adiciona nos vetores x e y
    saida+="Estado Atual: \n"
    saida+=str(estadoAtual)+"\n"
    saida+="\n\n"
    


    
    #Passo 1: Preve os valores x e y do próximo estado
    estadoPrevisto=preverProximoEstado(estadoAtual,roboQuadrupede)
    
    #Passo 2: Exibe os valores previstos
    XP=estadoPrevisto[0]
    YP=estadoPrevisto[1]
    
    saida+="Valores previstos: \n"
    saida+="XP= "+str(XP)+"\n"
    saida+="YP= "+str(YP)+"\n"
    saida+=("\n\n")
    
    #Passo 3: Propor uma ação de controle
    
    
    x0=torch.from_numpy(np.array([XP,YP,0,0])).float()
    yh0=controlador.forward(x0)
    
    CMP=(torchToNumpy(yh0)*360 -180)
    
    saida+="Ação de controle proposta: \n"
    saida+=str(CMP)+"\n\n\n"
    
    
    
    #Passo 4: Contruir uma entrada com a ação de controle proposta
    estadoCorrigido=construirEstadoComNovoComando(estadoAtual,CMP)
    
#     saida+="Estado alterado pelo controlador: \n"
#     saida+=str(estadoCorrigido)+" \n\n\n"
    
    
    #Passo 5: Utilizar a rede preditiva para verificar se o comando vai melhorar ou piorar o estado do robo
    
    estadoPrevisto=preverProximoEstado(estadoCorrigido,roboQuadrupede)
    
    XP_AposCorrecao=estadoPrevisto[0]
    YP_AposCorrecao=estadoPrevisto[1]
    
    saida+="Valores previstos após a correção: \n"
    saida+="XP= "+str(XP_AposCorrecao)+"\n"
    saida+="YP= "+str(YP_AposCorrecao)+"\n"
    saida+=("\n\n")
    
    
    #Passo 6: Se o estado melhorou, ele será aplicado no robo
    #         Se o estado piorou, o robo será penalizado
    
    erroAnterior=calcularErro(XP,YP,setPoint)
    erroAtual=calcularErro(XP_AposCorrecao,YP_AposCorrecao,setPoint)
    
    saida+=" Classificando ação de controle (RUIM/BOM):\n"
    
    
    
    if(np.abs(erroAnterior)<np.abs(erroAtual)):
        saida+="RUIM\n";
        usarCMP=False
        qErros+=1
#         controlador.penalizar()
        
    else:
        if(np.abs(erroAnterior)>np.abs(erroAtual)):
            saida+="BOM\n";
            usarCMP=True
            qAcertos+=1
#             controlador.recompensar()
        if(np.abs(erroAnterior)==np.abs(erroAtual)):
            saida+="Neutro\n";
            usarCMP=False
            qErros+=1


        
    saida+=("ERRO: ")
    saida+=str(erroAtual)
    saida+=("\n\n")
    saida+=("Acuracia:\n")
    saida+=str(qAcertos/(qAcertos+qErros))
    
    
    
    
    
    #Passo 7: Atualizar estado do robo
    
    if(usarCMP==False):
        xAtual=XP
        yAtual=YP
        estadoAtual=atualizarEstado(estadoAtual,CM,False,XP,XP)
        erro=erroAnterior
    else:
        xAtual=XP_AposCorrecao
        yAtual=XP_AposCorrecao
        estadoAtual=atualizarEstado(estadoAtual,CMP,False,XP_AposCorrecao,XP_AposCorrecao)
        erro=erroAtual
    
    
    X.append(xAtual)
    Y.append(yAtual)
    graficoErro.append(erro)
    
    saida+=("Tempo: "+str(1000*(time.time()-tic)//1)+"ms")
    print(saida)
    clear_output(wait=True)

    

```
